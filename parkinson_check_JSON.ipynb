```python
# Step 1: Define sensors and axes
sensors = ['Accelerometer', 'Gyroscope']
axes = ['X', 'Y', 'Z']

# Step 2: Generate new column names
new_columns = []
for record_name in df['record_name'].unique():
    for location in ['Left', 'Right']:
        for sensor in sensors:
            for axis in axes:
                new_columns.append(f"{record_name}_{location}_{sensor}_{axis}")

# Step 3: Create the transformed DataFrame
# Initialize with all new columns set to NaN
transformed_df = pd.DataFrame(columns=new_columns)

# Fill in values for the new structure
for _, row in df.iterrows():
    for sensor in sensors:
        for axis in axes:
            col_name_left = f"{row['record_name']}_Left_{sensor}_{axis}"
            col_name_right = f"{row['record_name']}_Right_{sensor}_{axis}"
            
            if sensor in row.index:
                if row['device_location'] == 'Left':
                    transformed_df.loc[row.name, col_name_left] = row[f"{sensor}_{axis}"]
                elif row['device_location'] == 'Right':
                    transformed_df.loc[row.name, col_name_right] = row[f"{sensor}_{axis}"]

# Reset index for clean output
transformed_df.reset_index(drop=True, inplace=True)

print(transformed_df.columns.tolist())
```
